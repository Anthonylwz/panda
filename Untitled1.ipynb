```python
#Crear series
serie=pd.Series([2,3,1,2,3,3.4], index=["a","b","c","d","e", "f"])
serie
```

```python
#Borrar seie o filas
```

```python
serie2=serie.drop(["a", "e"])
serie2
```

```python
#seleccionar filas o series
serie[["b","d"]]
```

```python
#operadores
serie4 = (serie + serie2)
serie4
```

```python
#resta
serie5 = (serie - serie2)
serie5
```

```python
#mutipicacion
serie6 = (serie * serie2)
serie6
```

```python
#division
serie7 = (serie / serie2)
serie7
```
